```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import arctan, sin

np.random.seed(0)

fun = lambda x: 4*arctan(x*np.pi)+sin(x*np.pi)
x_arr = np.asarray(np.random.sample(1000)*10)
y_arr = np.asarray([fun(x) for x in x_arr])

plt.plot(x_arr,y_arr, '*',color='red')
plt.show()





```

```python
from tensorflow.keras import models
from tensorflow.keras import layers

network = models.Sequential()
network.add(layers.Dense(40, activation='relu', input_dim=1))
network.add(layers.Dense(40))
network.add(layers.Dense(40))
network.add(layers.Dense(1))

network.compile(loss='mean_squared_error', optimizer='adam')

network.fit(x_arr, y_arr, epochs=1000, batch_size=124)

y_pred = network.predict(x_arr)
```

```python
plt.plot(x_arr,y_pred, 'o',color='blue')
plt.plot(x_arr,y_arr, '*',color='red')
plt.show()
```
